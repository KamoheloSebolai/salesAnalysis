```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salesData = pd.read_csv("Financial Sample.csv")
```

```python
# Convert currency columns to numeric values
currency_columns = ["Gross Sales", "Discounts", "Sales", "COGS", "Profit", "Manufacturing Price", "Sale Price"]

# remove leading and trailing spaces
salesData.columns = salesData.columns.str.strip()

for column in currency_columns:
        salesData[column] = salesData[column].replace(r'[\$,]', '', regex=True).astype(float)

salesData
```

```python
# Bar plot of total units sold by country
plt.figure(figsize=(8, 6))
sns.barplot(x='Country', y='Units Sold', data=salesData)
plt.title('Total Units Sold by Country')
plt.show
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Country', y='Profit', data=salesData)
plt.title('Total Profit by Country')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Profit', hue='Country', data=salesData)
plt.title('Profit Trend by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Profit', data=salesData)
plt.title('Average Profit by Country')
plt.show()
```
