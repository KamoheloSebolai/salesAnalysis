```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salesData = pd.read_csv("Financial Sample.csv")
```

```python
# Convert currency columns to numeric values
currency_columns = ["Gross Sales", "Discounts", "Sales", "COGS", "Profit", "Manufacturing Price", "Sale Price"]

# remove leading and trailing spaces
salesData.columns = salesData.columns.str.strip()

for column in currency_columns:
        salesData[column] = salesData[column].replace(r'[\$,]', '', regex=True).astype(float)

salesData
```

```python
# Summary statistics
summary = {
    "Total Units Sold": salesData["Units Sold"].sum(),
    "Total Gross Sales": salesData["Gross Sales"].sum(),
    "Total Discounts": salesData["Discounts"].sum(),
    "Total Sales": salesData["Sales"].sum(),
    "Total COGS": salesData["COGS"].sum(),
    "Total Profit": salesData["Profit"].sum(),
    "Average Manufacturing Price": salesData["Manufacturing Price"].mean(),
    "Average Sale Price": salesData["Sale Price"].mean()
}

print("Summary Statistics:")
for key, value in summary.items():
    print(f"{key}: ${float(value):,.2f}")
```

```python
# Sales Trend Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Sales', data=salesData)
plt.title('Sales Trend Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
discount_impact = salesData.groupby('Discount Band').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

plt.figure(figsize=(12, 6))
discount_Band_melted = discount_impact.melt(id_vars='Discount Band', value_vars=['Sales', 'Profit'])
sns.barplot(x='Discount Band', y='value', hue='variable', data=discount_Band_melted)
plt.title('Discount Impact on Sales and Profit')
plt.figtext(0.5, -0.12, 
            "This chart highlights the complex relationship between discounts, sales, and profit. "
            "While discounts can be a powerful tool to attract customers, they need to be used strategically "
            "to maximize profitability.", 
            wrap=True, horizontalalignment='center', fontsize=12, color='black')
plt.show()
```
